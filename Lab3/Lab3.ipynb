```python
# 4) Perform the following data exploration analysis on the downloaded dataset
# a) Count plot
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("C:\\Users\\Ananya\\PycharmProjects\\BottleFilling\\Documents\\Iphone data.csv")
sns.countplot(x='Discount Percentage', data=df)  # The hue parameter allows you to specify a categorical variable to group the data by and create separate counts for each category.
plt.show()
```

```python
# # d) Correlation
from scipy.stats import pearsonr
def get_correlation(column1, column2, df):
    df.dropna(inplace=True)
    pearson_corr, p_value = pearsonr(df[column1], df[column2])
    print("Correlation between {} and {} is {}".format(column1, column2, pearson_corr))
    print("P-value of this correlation is {}".format(p_value))
get_correlation('Sale Price','Discount Percentage', df)
```
